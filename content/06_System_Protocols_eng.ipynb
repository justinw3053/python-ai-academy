### 🎯 Μάθημα 34: 🛡️ Προστασία από Σφάλματα (Try / Except)

**Η Ιστορία:** Μερικές φορές ο κώδικάς μας μπορεί να κάνει ένα λάθος (όπως διαίρεση με το μηδέν) και το πρόγραμμα να σταματήσει απότομα (crash). Στην Python μπορούμε να 'πιάσουμε' αυτά τα σφάλματα με τις εντολές `try` και `except` για να μην κρασάρει το παιχνίδι μας!

**Στόχος:** Wrap-αρε μια επίμονη διαίρεση `10 / 0` μέσα σε ένα block `try` και `except ZeroDivisionError:` και εμφάνισε το μήνυμα `'Σφάλμα: Διαίρεση με το μηδέν!'`.

### 🎯 Μάθημα 36: ✍️ Εγγραφή σε Αρχεία (with open - "w")

**Η Ιστορία:** Εκτός από το να διαβάζουμε, μπορούμε και να γράφουμε νέα αρχεία στο δίσκο! Αν ανοίξεις ένα αρχείο χρησιμοποιώντας τη λειτουργία `"w"` (Write), η Python θα δημιουργήσει το αρχείο (ή θα σβήσει το παλιό και θα γράψει από πάνω) και θα γράψει ό,τι της ζητήσεις.

**Στόχος:** Δημιούργησε ένα αρχείο με όνομα `log.txt` και γράψε μέσα τη φράση `'ΣΥΣΤΗΜΑΤΑ ΑΣΦΑΛΗ'` χρησιμοποιώντας τη μέθοδο `.write()`.


```python
# 💻 Άνοιξε το log.txt με λειτουργία "w" και γράψε μέσα 'ΣΥΣΤΗΜΑΤΑ ΑΣΦΑΛΗ':
# (Συμπλήρωσε τα κενά από κάτω!)

with open("log.txt", "___", encoding="utf-8") as file:  # Βάλε τη λειτουργία εγγραφής "w"
    file.write("___")  # Γράψε τη φράση 'ΣΥΣΤΗΜΑΤΑ ΑΣΦΑΛΗ'


# === AUTOMATED CHECK ===
try:
    import os
    assert os.path.exists("log.txt"), "Το αρχείο 'log.txt' δεν δημιουργήθηκε!"
    with open("log.txt", "r", encoding="utf-8") as f:
        log_content = f.read().strip()
    assert "ΣΥΣΤΗΜΑΤΑ ΑΣΦΑΛΗ" in log_content, "Το log.txt δεν περιέχει το σωστό μήνυμα!"
    print("\n🎉 [SYSTEM] Great job! Έγραψες με επιτυχία το πρώτο σου αρχείο log.txt!")
    print("👉 Type 'progress' to the Pi Agent για να δει τι κατάφερες!")
    
    # Καταγραφή Προόδου
    try:
        import sys
        sys.path.append("/home/justin/Space_Station_Academy/.pi")
        import tracker
        tracker.update_progress("Mission 6, Μάθημα 36: Writing Files")
    except Exception:
        pass
except AssertionError as e:
    print("\n🤖 [Pi Agent] Don't be discouraged! Type 'Help' in the side window so we can solve it together!")
    print(f"\n❌ [SYSTEM] Error: {e}")
except Exception as e:
    print("\n🤖 [Pi Agent] Something went wrong with the code. Type 'Help' so we can look at it!")
    print(f"\n❌ [SYSTEM] Error: {e}")
```

### 📦 The Secret of the Box (Indentation)!

Before we start making decisions, you need to know something critical! In Python, when we want to put commands *inside* a decision (an `if` statement), we use **Indentation** (by pressing the **Tab** key on your keyboard, or 4 spaces).

Think of it like a **Box**: Whatever is pushed further inside (has empty space on the left) is *inside* the `if` box!
